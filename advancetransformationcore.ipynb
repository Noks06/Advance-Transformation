```python
# Importing relevent liberies 
## Standard Imports
import numpy as np
import pandas as pd
## Importing the OS and JSON Modules
import os,json
df = pd.read_csv('/Users/noksmchunu/Downloads/superhero_info - superhero_info.csv')
df.head()

```

```python
# exploring info
df.info()
```

```python
## save the 2 new columns( hero and publisher) into the dataframe
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head(2)
```

```python
## drop the original column 
df = df.drop(columns=['Hero|Publisher'])
df.head(2)
```

```python
## use .str.replace to replace all single quotes
## examining a single value from the coordinates col
Measure = df.loc[0,"Measurements"]
print(type(Measure))
Measure
```

```python
import json
Measure = Measure.replace("'",'"')
Measure

json.loads(Measure)
```

```python
## use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
# Separating weight and height column
Hight_Weight = df['Measurements'].apply(pd.Series)
Hight_Weight
```

```python
# concat long_lat with original dataframe
df = pd.concat((df, Hight_Weight), axis = 1)
df.head(2)
```

```python
# Make a list of all characters to replace
to_replace = (['cm','kg'])
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    df['Height'] = df['Height'].str.replace(char,'',regex=False)
    df['Weight'] = df['Weight'].str.replace(char,'',regex=False)
df.head()
```

```python
# drop the Measure column 
df = df.drop(columns=['Measurements'])
df.head(2)
```

```python
df_powers = pd.read_csv('/Users/noksmchunu/Downloads/superhero_powers - superhero_powers.csv')
df_powers.head()
```

```python
# concat long_lat with original dataframe
df = pd.concat((df, df_powers['Powers']), axis = 1)
df.head(5)

```

```python
## showing the lists are really strings
(df.loc[2,'Powers'])
df.head()
```

```python
#change the weight and Height into Float data type  
df['Weight'] = df['Weight'].astype(float)
df['Height'] = df['Height'].astype(float)
```

```python
publisher_weight  = df[['Publisher','Weight', 'Height']]
publisher_weight.head()
```

```python
df['Powers'].value_counts()
```

```python
## exploding the column of lists
exploded = df.explode('Powers')
exploded[['Hero','Powers','Powers']].head()
```

```python
## saving the unique values from the exploded column
cols_to_make = df['Powers'].str.split(',').explode().unique().tolist()
cols_to_make
```

```python
pd.set_option('display.max_columns', None)
```

```python
for col in cols_to_make:
    df[col] = df['Powers'].str.contains(col)
df.head()
```

Compare the average weight of super powers who have Super Speed to those who do not.



```python
super_speed =df.loc[df['Super Speed']]
super_speed
```

```python
avg_weight_super = super_speed.mean(numeric_only =True)
avg_weight_super['Weight']
```

```python
notsuper_speed =df.loc[~df['Super Speed']]
notsuper_speed
```

```python
avg_weight_notsuper = notsuper_speed.mean(numeric_only =True)
avg_weight_notsuper['Weight']
```
